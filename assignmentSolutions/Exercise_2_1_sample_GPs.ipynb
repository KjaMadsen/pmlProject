```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(24)
```

## Exercise 1

```python
num_of_points = 101
S = np.linspace(0, 1, num_of_points)
```

```python
def k_rbf(x, y, l): # l = gamma
    return np.exp(-l*(x-y)**2)
```

```python
def get_K(kernel, x, l):
    N = len(x)
    M = np.zeros((N, N))
    for i in range(N):
        for j in range(i+1,N):
            M[i,j] = kernel(x[i], x[j], l)
    K = M + M.T
    np.fill_diagonal(K, 1)
    return(K)
```

```python
num_samples = 5
K = get_K(k_rbf, S, 10)
prior_samples = np.random.multivariate_normal(mean=np.zeros(num_of_points), cov=K, size=num_samples)
```

```python
for i in range(num_samples):
    plt.plot(S, prior_samples[i,:])
plt.show()
```
